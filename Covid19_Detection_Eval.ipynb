```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.applications import VGG16

# Define the architecture (Must match the training notebook)
base_model = VGG16(weights=None, include_top=False, input_shape=(224, 224, 3))
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dense(2, activation='softmax')
])

# Path to your saved weights (Update path based on your Kaggle Input location)
# Example: '/kaggle/input/your-dataset-name/best_model_weights.weights.h5'
weights_path = '/kaggle/input/datasets/usmansarwar143/dl-results/working/best_model_weights.weights.h5' 
model.load_weights(weights_path)

print("Model architecture reconstructed and weights loaded successfully.")
```

```python
def prepare_image(image_path):
    target_size = (224, 224)
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, target_size)
    image = image / 255.0  # Normalization
    return np.expand_dims(image, axis=0) # Add batch dimension

labels = {0: 'non-COVID', 1: 'COVID'}
```

```python
# Replace with a real path from your dataset
test_img_path = '/kaggle/input/datasets/plameneduardo/sarscov2-ctscan-dataset/COVID/Covid (1000).png'

if os.path.exists(test_img_path):
    prepared_img = prepare_image(test_img_path)
    predictions = model.predict(prepared_img)
    predicted_class = np.argmax(predictions)
    confidence = np.max(predictions) * 100

    plt.imshow(cv2.cvtColor(cv2.imread(test_img_path), cv2.COLOR_BGR2RGB))
    plt.title(f"Prediction: {labels[predicted_class]} ({confidence:.2f}%)")
    plt.axis('off')
    plt.show()
else:
    print("Image path not found. Please verify the dataset path.")
```
